```python
 df_combined_wine.describe().pH
```

```python
bins = [2.72,3.11,3.21,3.32,4.01]
bin_names = ['very-high-pH','high-pH','medium-pH','low-pH']
df_combined_wine['acidity_levels'] = pd.cut(df_combined_wine['pH'],bins, labels=bin_names)
```

```python
df_combined_wine
```

```python
df1 = df_combined_wine.groupby('acidity_levels').mean().quality
df1
```

```python
# Create a bar chart with proper labels
locations = [1, 2,3,4]
heights = df1
labels = ['very-high-pH','high-pH','medium-pH','low-pH']
plt.bar(locations, heights, tick_label=labels)
plt.title('Average Quality Ratings by pH Level')
plt.xlabel('pH Level')
plt.ylabel('Average Quality Rating');
```

```python
### Higher the pH levels, quality rating is lower.
```

```python
### quality of wines with high and low alcohol ?
```

```python
alcohol_mean=df_combined_wine['alcohol'].mean()
```

```python
df_low_alcohol = df_combined_wine.query('alcohol < {}'.format(alcohol_mean))
```

```python
df_high_alcohol = df_combined_wine.query('alcohol > {}'.format(alcohol_mean))
```

```python
mean_quality_low_alcohol = df_low_alcohol['quality'].mean()
mean_quality_high_alcohol = df_high_alcohol['quality'].mean()
```

```python
locations = [1, 2]
heights = [mean_quality_low_alcohol,mean_quality_high_alcohol]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels)
plt.title('Avg quality by alcohol content')
plt.xlabel('Quality')
plt.ylabel('Alcohol content')
```

```python
### High alcohol content has been marked with high quality
```
